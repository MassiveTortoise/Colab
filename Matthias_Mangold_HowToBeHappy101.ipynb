```python
mydata['Ladder score'] = pd.to_numeric(mydata['Ladder score'], errors='coerce')
q1 = mydata['Ladder score'].quantile(0.25)
q3 = mydata['Ladder score'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Find outliers
outliers = mydata[(mydata['Ladder score'] < lower_bound) | (mydata['Ladder score'] > upper_bound)]
outliers
```
